Menentukan nilai maksimum/minimum dari suatu data

```python
import pandas as pd
df = pd.read_csv("H05.csv")


imax = df["nilai_mat"].idxmax()
print(df[imax:imax+1])

imin = df["nilai_mat"].idxmin()
print(df[imin:imin+1])
```

Mengurutkan Data

```python
print(df.sort_values(["nilai_mat", "nilai_kim"], ascending=[0,1]))
```

```python
print(df["nilai_mat"].value_counts())
```

```python
minimum = df.min()
maximum = df.max()

print(maximum["nilai_mat"])
print(minimum["fakultas"])
```

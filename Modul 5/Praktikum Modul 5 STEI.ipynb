PROBLEM 3

```python
#Histogram distribusi banyaknya review di aplikasi yang memiliki jenis penyediaan ruang Private room dengan bins 30.
df[df['room_type'] == 'Private room']['number_of_reviews'].plot(kind='hist', bins=30, rwidth=0.9,title='Distribusi Review Aplikasi pada Ruangan tipe Private Room')
```

```python
#3. Pie chart perbandingan banyaknya masing-masing jenis penyediaan ruang dengan banyaknya review di aplikasi di atas 100.
df[df['number_of_reviews']>100]['room_type'].value_counts().plot(kind='pie', title='Distribusi Jenis Penyedian Ruang pada Rumah dengan Review di atas 100', ylabel='')
```

```python
#3. Scatter plot dengan long sebagai x dan lat sebagai y dari rumah yang tidak dimiliki oleh Jay.
df[df['host_name'] != 'Jay'].plot(kind='scatter', x='longitude', y='latitude',title='Hubungan Longitude v Latitude dari Rumah yang Tidak Dimiliki Jay')
```

```python
#4. Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya hari ketersediaan rumah dalam satu tahun sebagai ukuran dari rumah yang berada di neighbourhood Novena.
df[df['neighbourhood'] == 'Novena'].plot(kind='scatter', x='longitude', y='latitude', sizes=df['availability_365'], title='Banyaknya hari ketersediaan rumah dalam satu tahun di Perumahan Novena',color='red')
```

```python
#5. Stacked bar chart dari rumah yang disewakan oleh ’JoJo’, ’Elizabeth’, dan ’Jimmy’, dengan room type sebagai x dan nama pemilik rumah sebagai stacked y.
df[(df['host_name']=='Jojo')| (df['host_name']=='Elizabeth') | (df['host_name']=='Jimmy')].groupby(['room_type','host_name']).size().unstack().plot(kind='bar', stacked=True)
```
